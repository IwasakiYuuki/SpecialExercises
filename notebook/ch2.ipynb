```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise2.2

```python
class FormalNeuronLayer:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def forward(self, x):
        p = np.dot(self.w, x)
        y = p > self.h
        return y.astype(np.int)
```

```python
x = np.array([[0,0,1,1],
              [0,1,0,1]])
w = np.array([[1, 1],
              [-1, -1]])
h = np.array([[0],
              [-2]])
u = np.array([1, 1])
g = np.array([1])

layer1 = FormalNeuronLayer(w, h)
layer2 = FormalNeuronLayer(u, g)

y = layer1.forward(x)
z = layer2.forward(y)
print(y)
print(z)
```

## Exercise2.3

```python
class Affine:
    def __init__(self, w, b):
        self.w = w
        self.b = b
    
    def forward(self, x):
        p = np.dot(self.w, x) + self.b
        return p
    
class Step:
    def forward(self, x):
        y = x > 0
        return y.astype(np.int)
```

```python
x = np.array([[0,0,1,1],
              [0,1,0,1]])
w = np.array([[1, 1],
              [-1, -1]])
b = np.array([[0],
              [2]])
u = np.array([1, 1])
c = np.array([-1])

layer1 = Affine(w,b)
layer2 = Step()
layer3 = Affine(u,c)
layer4 = Step()

p = layer1.forward(x)
y = layer2.forward(p)
q = layer3.forward(y)
z = layer4.forward(q)
print(y)
print(z)
```

## Exercise2.4

```python
class Sigmoid:
    def forward(self, x):
        return 1 / (1 + np.exp(-x))
```

```python
x = np.array([[0,0,1,1],
              [0,1,0,1]])
w = np.array([[1, 1],
              [-1, -1]])
b = np.array([[0],
              [2]])
u = np.array([1, 1])
c = np.array([-1])

layer1 = Affine(w,b)
layer2 = Sigmoid()
layer3 = Affine(u,c)
layer4 = Sigmoid()

p = layer1.forward(x)
y = layer2.forward(p)
q = layer3.forward(y)
z = layer4.forward(q)
print(y)
print(z)
```
